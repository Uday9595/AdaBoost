```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
data = pd.concat([train,test], sort= False)
data.info()
```

```python
data.isnull()
```

```python
data['Age']= data['Age'].fillna(value=data['Age'].median())
data['Fare']= data['Fare'].fillna(value=data['Fare'].median())
```

```python
data.info()
```

```python
data = data.reset_index()
```

```python
sns.catplot(x = 'Embarked', kind = 'count', data = data)
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
data.info()
```

```python
data.loc[ data['Age'] <= 16, 'Age'] = 0
data.loc[(data['Age'] > 16) & (data['Age'] <= 32), 'Age'] = 1
data.loc[(data['Age'] > 32) & (data['Age'] <= 48), 'Age'] = 2
data.loc[(data['Age'] > 48) & (data['Age'] <= 64), 'Age'] = 3
data.loc[ data['Age'] > 64, 'Age'] = 4
```

```python
import re
def get_title(name):
    title_search = re.search(' ([A-Za-z]+\.)', name)

    if title_search:
        return title_search.group(1)
    return ""
```

```python
data['Title'] = data['Name'].apply(get_title)
data['Title'].value_counts()
```

```python
data['Title'] = data['Title'].replace(['Capt.', 'Dr.', 'Major.', 'Rev.'], 'Officer.')
data['Title'] = data['Title'].replace(['Lady.', 'Countess.', 'Don.', 'Sir.', 'Jonkheer.', 'Dona.'], 'Royal.')
data['Title'] = data['Title'].replace(['Mlle.', 'Ms.'], 'Miss.')
data['Title'] = data['Title'].replace(['Mme.'], 'Mrs.')
data['Title'].value_counts()
```

```python
data['Cabin'] = data['Cabin'].fillna('Missing')
data['Cabin']= data['Cabin'].str[0]
data['Cabin'].value_counts()
```

```python
data['Family_Size'] = data['SibSp'] + data['Parch'] + 1
data['IsAlone'] = 0
data.loc[data['Family_Size']==1, 'IsAlone'] = 1
data.head()
```

```python
data1 = data.drop(['Name','Ticket'],axis =1)
```

```python
data1
```

```python
datadummies = pd.get_dummies(data, drop_first = True)
datadummies.head()
```

```python
datatrain = datadummies[datadummies['Survived'].notna()]
datatrain.info()
```

```python
datatest= datadummies[datadummies['Survived'].isna()]
datatest.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(datatrain.drop(['PassengerId','Survived'],axis=1),
                                                    datatrain['Survived'], test_size=0.30,
                                                    random_state=101, stratify = datatrain['Survived'])
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
ada= AdaBoostClassifier(DecisionTreeClassifier(),n_estimators =100, random_state= 0)
ada.fit(X_train,y_train)
```

```python
predictions = ada.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
print(f'Train Accuracy - :{ada.score(X_train,y_train):.3f}')
print(f'Test Accuracy - :{ada.score(X_test,y_test):.3f}')
```
